# Make Circles Example With Noise

```python
from dim_reduction import KernelPCA
from datasets import make_circles, print_shape
from kernels import RBFKernel, PolynomialKernel, LinearKernel
from plots import scatter2D, scatter3D
```

```python
X, y = make_circles(n_samples=500, n_features=3, n_circles=3, balanced=False, noise=0.01, random_state=65)
```

```python
print_shape(X, y)
```

```python
title = "2D Original Make Circles Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
scatter2D(title, xlabel, ylabel, X, y)
```

```python
title = "3D Original Make Circles Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
zlabel = "Feature 3"
scatter3D(title, xlabel, ylabel, zlabel, X, y)
```

```python
kernel = LinearKernel()
pca = KernelPCA(kernel = kernel, n_components=2)
X_trans = pca.fit_transform(X)
```

```python
print_shape(X, y)
```

```python
title = "2D Linear Kernel PCA Make Circles Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
scatter2D(title, xlabel, ylabel, X_trans, y)
```

```python
kernel = LinearKernel()
pca = KernelPCA(kernel = kernel, n_components=3)
X_trans = pca.fit_transform(X)
```

```python
print_shape(X, y)
```

```python
title = "3D Linear Kernel PCA Make Circles Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
zlabel = "Feature 3"
scatter3D(title, xlabel, ylabel, zlabel, X_trans, y)
```

```python
kernel = RBFKernel(gamma=1.0)
pca = KernelPCA(kernel = kernel, n_components=2)
X_trans = pca.fit_transform(X)
```

```python
print_shape(X, y)
```

```python
title = "2D RBF Kernel PCA Make Circles Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
scatter2D(title, xlabel, ylabel, X_trans, y)
```

```python
kernel = RBFKernel(gamma=1.0)
pca = KernelPCA(kernel = kernel, n_components=3)
X_trans = pca.fit_transform(X)
```

```python
print_shape(X, y)
```

```python
title = "3D RBF Kernel PCA Make Circles Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
zlabel = "Feature 3"
scatter3D(title, xlabel, ylabel, zlabel, X_trans, y)
```

```python
kernel = PolynomialKernel(degree=4)
pca = KernelPCA(kernel = kernel, n_components=2)
X_trans = pca.fit_transform(X)
```

```python
print_shape(X, y)
```

```python
title = "2D Polynomial Kernel PCA Make Circles Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
scatter2D(title, xlabel, ylabel, X_trans, y)
```

```python
kernel = PolynomialKernel(degree=4)
pca = KernelPCA(kernel = kernel, n_components=3)
X_trans = pca.fit_transform(X)
```
